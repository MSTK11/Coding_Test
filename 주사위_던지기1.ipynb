```python
N, M = map(int, input().split())
```

```python
import copy
def dice_1(n):
  if n == N:
    case_1 = copy.deepcopy(case)
    result.append(case_1)
    return
  for i in range(1, 7):
    case[n] = i
    dice_1(n+1)
```

```python
def dice_2(n):
  if n == N:
    case_1 = copy.deepcopy(case)
    case_1.sort()
    if case_1 not in result:
      result.append(case_1)
    return
  for i in range(1, 7):
    case[n] = i
    dice_2(n+1)
```

```python
def dice_3(n):
  if n == N:
    case_1 = copy.deepcopy(case)
    if len(set(case_1)) == N:
      result.append(case_1)
    return
  for i in range(1, 7):
    case[n] = i
    dice_3(n+1)
```

```python
if __name__ == "__main__":
  case = [0] * N
  result = []
  if M == 1:
    dice_1(0)
    for i in result:
      print(*i)
  elif M == 2:
    dice_2(0)
    for i in result:
      print(*i)
  elif M == 3:
    dice_3(0)
    for i in result:
      print(*i)
```
